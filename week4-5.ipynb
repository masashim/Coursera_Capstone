1) Scraping “List of cities by GDP” data from Wikipedia web page, putting into Dataframe. 

2) Deleting no GDP data lows from Dataframe to proceed.

```python
import pandas as pd

dfs = pd.read_html("https://en.wikipedia.org/wiki/List_of_cities_by_GDP", header=[0])
df = dfs[0]
df = df.dropna(how="any")
df = df.astype({'Brookings Institution[3]2014 est.PPP-adjustedGDP ($BN)':'float64', 'PwC[4]2008 est.PPP-adjustedGDP ($BN)' : 'float64', 'McKinsey[5]2010 est.NominalGDP ($BN)' : 'float64'})
# print(df.dtypes)
df = df[['City proper /Metropolitan area', 'Country/Region', 'PwC[4]2008 est.PPP-adjustedGDP ($BN)', 'McKinsey[5]2010 est.NominalGDP ($BN)', 'Brookings Institution[3]2014 est.PPP-adjustedGDP ($BN)']]
df
# print(df.applymap(type))
```

3) Sorting cities by using GDP number.

```python
df_sort = df.sort_values('Brookings Institution[3]2014 est.PPP-adjustedGDP ($BN)', ascending=False)
pd.set_option('display.max_rows', None)
df_sort
```

4) Displaying bar charts to make sure priorities by GDP numbers.

```python
df_sort.plot.bar(x = 'City proper /Metropolitan area')
```

```python
import geocoder
import folium

# Initializing map, setting map center, zoom start
map = folium.Map(location=geocoder.arcgis('Malta').latlng, zoom_start=2)

prefecture_list = []
# Setting circle size weight
WEIGHT = 0.07
for index, row in df.iterrows():
    prefecture = row['City proper /Metropolitan area'] 
    value = row['Brookings Institution[3]2014 est.PPP-adjustedGDP ($BN)'] 
    value2 = row['PwC[4]2008 est.PPP-adjustedGDP ($BN)']
    value3 = row['McKinsey[5]2010 est.NominalGDP ($BN)']
# Getting Latitude and Longitude from city names
    g = geocoder.arcgis(prefecture)   
    prefecture_list.append(g.latlng)
# Putting circlemaker on map by using Latitude and Longitude from city names
    folium.CircleMarker(g.latlng, radius=value * WEIGHT,color='blue',fill_color='blue').add_to(map) 
    folium.CircleMarker(g.latlng, radius=value2 * WEIGHT,color='blue',fill_color='blue').add_to(map) 
    folium.CircleMarker(g.latlng, radius=value3 * WEIGHT,color='blue',fill_color='blue').add_to(map) 
# Displaying map
map
```

```python
from folium.plugins import HeatMap

# Initializing map, setting map center, zoom start
map = folium.Map(location=geocoder.arcgis('Malta').latlng, zoom_start=2)

prefecture_list = []
for index, row in df.iterrows():
    prefecture = row['City proper /Metropolitan area'] 
# Getting Latitude and Longitude from city names
    g = geocoder.arcgis(prefecture)   
    prefecture_list.append(g.latlng)
# Creating heatmap by using using Latitude and Longitude from city names
HeatMap(prefecture_list).add_to(map)
# Displaying map
map
```
